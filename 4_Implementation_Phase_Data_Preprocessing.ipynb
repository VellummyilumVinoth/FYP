````python
# Import the required libraries
import re
import os
import pandas as pd
````

````python
file_path = '/content/drive/MyDrive/Final FYP/transformed_dataset_with_code_snippets.csv'
dataset = pd.read_csv(file_path)
````

````python
dataset
````

````python
# Assuming dataset is your DataFrame
dataset = dataset.rename(columns={'subproblem': 'prompt', 'code_snippet': 'response'})
````

````python
# Inspect the modified dataset
dataset.head(5)
````

````python
# Define a function to extract Java code snippets
def extract_java_code(text):
    match = re.search(r'```java(.*?)```', text, re.DOTALL)
    if match:
        return match.group(1).strip()
    return None

# Apply the function to the 'code_snippet' column
dataset['extracted_code'] = dataset['response'].apply(extract_java_code)

# Drop rows with no extracted code
dataset = dataset.dropna(subset=['extracted_code'])
````

````python
# Keep only the 'prompt' and 'extracted_code' columns
dataset = dataset[['prompt', 'extracted_code']]
````

````python
# Rename the columns
dataset.rename(columns={'extracted_code': 'response'}, inplace=True)
````

````python
dataset.head()
````

````python
# Define the path to save the dataset
save_path = '/content/drive/My Drive/Final FYP/preprocessed_codes_snippet_dataset.csv'

# Save the dataset to the specified path
dataset.to_csv(save_path, index=False)

# Print a confirmation message
print(f"Dataset saved to {save_path}")
````
